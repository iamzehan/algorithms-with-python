#Maximize it!

```python
import itertools

def maximize_it(arrays,M):
    max_modulo = 0
    for product in itertools.product(*arrays):
        modulo = sum(x**2 for x in product)%M
        if modulo > max_modulo:
            max_modulo = modulo
    return max_modulo

if __name__ == '__main__':
    k, M = input().split()
    k, M = int(k),int(M)
    arrays=[]
    for i in range(k):
        input_string = input()
        array = [int(x) for x in input_string.split()]
        size = array.pop(0)
        if int(size)==len(array):
            arrays.append(array)
    print(maximize_it(arrays,M))
```
